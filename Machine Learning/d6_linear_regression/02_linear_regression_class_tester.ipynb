# Let's test your implementation of LinearRegression 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Import your class here: Deep note not letting more import 
class LinearRegression:

    def __init__(self):
        self.a = 0
        self.b = 0

    def fit(self, x, y):
        # determine what the a and b are 
        x_mean = np.mean(x)
        y_mean = np.mean(y)
        self.b = sum((x-x_mean)*(y-y_mean))/sum(np.square(x-x_mean))
        self.a = y_mean - self.b * x_mean
        return self
        
    def predict(self, x_pred):
        y = self.a + self.b * x_pred
        return y

    def plot(self,x,y):
        sns.scatterplot(x, y)
        sns.lineplot(x=x, y=(self.a+self.b*x), color='red')
        plt.show()
```

```python
assert LinearRegression is not None, 'You should import your class'
assert 'fit' in dir(LinearRegression) and 'predict' in dir(LinearRegression), 'You should follow sklearn conventions'
```

```python
# Load the data in X and y
df = pd.read_csv('reg_data.csv')

X=df.iloc[:,0].values
y=df.iloc[:,1].values
```

```python
assert X is not None and type(X) == np.ndarray, 'X should be defined and be a numpy array'
assert y is not None and type(y) == np.ndarray, 'y should be defined and be a numpy array'
```
