#### Math Background
The task is to solve linear regression for the data set of ```x``` and ```y```. That is, fit a line y to the data such that ```y = a + bx + e```. Where a and b are coefficents and e is an error term.

We know that ```b = SUM ( xi - x_mean) * (yi - y_mean) / SUM (xi - x_mean)^2``` where ```xi```, and ```yi``` are the indivdual observations and ```x_mean```, ```y_mean``` are means of all ```xi``` and ```yi```.

So a reasonable solution then for a fit is ```a = y_mean - b * x_mean```.

````python
#import the minimum packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
#load and inspect the data
df = pd.read_csv('reg_data.csv')
df.head()
````

````python
#plot the data
x=df.iloc[:,0]
y=df.iloc[:,1]
sns.scatterplot(x, y)
````

Write your inital observations below:



````python
#implement a least squares function to find a, b
# b = SUM ( xi - x_mean) * (yi - y_mean) / SUM (xi - x_mean)^2
# a = y_mean - b * x_mean
x_mean = np.mean(x)
y_mean = np.mean(y)
b = sum((x-x_mean)*(y-y_mean))/sum(np.square(x-x_mean))
a = y_mean - b * x_mean
print(b, a)
````

````python
#plot the regression line as x and y
sns.lineplot(x=x, y=(a+b*x))
plt.show()
````

````python
#plot the original data with the regression line as defined by your function
sns.scatterplot(x, y)
sns.lineplot(x=x, y=(a+b*x), color='red')
plt.show()
````

Easy enough to do in the end. I have created it a class below. 

````python
class LinearRegression:

    def __init__(self):
        self.a = 0
        self.b = 0

    def fit(self, x, y):
        # determine what the a and b are 
        x_mean = np.mean(x)
        y_mean = np.mean(y)
        self.b = sum((x-x_mean)*(y-y_mean))/sum(np.square(x-x_mean))
        self.a = y_mean - self.b * x_mean
        return self
    def predict(self, x_pred):
        y = a + b * x_pred
        return y

    def plot(self,x,y):
        sns.scatterplot(x, y)
        sns.lineplot(x=x, y=(self.a+self.b*x), color='red')
        plt.show()
````

````python
lr = LinearRegression()

lr.plot(x,y)
````
